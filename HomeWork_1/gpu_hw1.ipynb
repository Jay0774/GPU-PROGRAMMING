```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('gpu.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10,10),dpi = 100)
plt.plot(df['Size of Array'],df[['Time Sequential (ms)','Time Vector (ms)','Time Parallel (ms)']])
plt.title('Vector, Parallel(manual), Sequential')
plt.xlabel('Size of Array')
plt.ylabel('Time (ms)')
plt.legend(['Sequential','Vector','Parallel'])
```

```python
plt.figure(figsize=(10,10),dpi = 100)
plt.plot(df['Size of Array'],df[['Time Sequential (ms).2','Time vector (ms)','Time Parallel (ms).2']])
plt.title('Vector, Parallel(compiler), Sequential')
plt.xlabel('Size of Array')
plt.ylabel('Time (ms)')
plt.legend(['Sequential','Vector','Parallel'])
```

```python
plt.figure(figsize=(10,10),dpi = 100)
plt.plot(df['Size of Array'],df[['Time Sequential (ms)','Time Sequential (ms).2']])
plt.title('Sequential Programming')
plt.xlabel('Size of Array')
plt.ylabel('Time (ms)')
plt.legend(['Parallel (manual)','Parallel (Compiler)'])
```

```python
plt.figure(figsize=(10,10),dpi = 100)
plt.plot(df['Size of Array'],df[['Time Parallel (ms)','Time Parallel (ms).2']])
plt.title('Parallel Programming')
plt.xlabel('Size of Array')
plt.ylabel('Time (ms)')
plt.legend(['Parallel (manual)','Parallel (Compiler)'])
```

```python
plt.figure(figsize=(10,10),dpi = 100)
plt.plot(df['Size of Array'],df[['Time Vector (ms)','Time vector (ms)']])
plt.title('Vector Programming')
plt.xlabel('Size of Array')
plt.ylabel('Time (ms)')
plt.legend(['Parallel (manual)','Parallel (Compiler)'])
```
